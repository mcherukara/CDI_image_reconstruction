```python
import tifffile as tff
import numpy as np
import scipy.fftpack as sf
import matplotlib.pyplot as plt
import math as m
import Tools as T
from scipy.ndimage import center_of_mass as com
%matplotlib inline
```

# Some phasing parameters

```python
#Input file name
ifile='summededit1crop.tif'

#Number of HIO, ER, OSS and cycles
n_HIO=90
n_ER=20
n_OSS=0
n_cycles=100 #Number of times to cycle ER+HIO+ER
n_iters=(2*n_ER+n_HIO+n_OSS)*n_cycles

#Shrink wrap variables (Gaussian smoothing)
isig=5 #Sigma of G for first support update
fsig=1 #Sigma for final support update
severy=5 #Update support every how many iterations?
sfrac=0.15 #Fraction of maximum intensity to use for shrink wrap boundary

#Phasing threshold and image subtraction threshold
thresh=10

#Phase range
nph=-np.pi/2
pph=np.pi/2

#Feedback parameter (set between 0.7 and 0.9)
beta=0.9

#Data array size is how many times image array?
dfs=12/10.

#Initial box support (in fractions of data array size)
xf=0.4
yf=0.4
zf=0.5
```

# Some functions

```python
#ER
def ER(dmag,data,support,guess,err,merr):
  iguess=sf.fftn(guess)
  cerr=np.sum( ( np.abs(iguess)-np.abs(data) )**2 )/dmag #Calculate error
  phi=np.arctan2(iguess.imag,iguess.real)
  iguess=np.where(np.abs(data),data*np.exp(1.j*phi),0.0) #Replace with measured amplitudes
  oguess=guess
  guess=sf.ifftn(iguess)
  sos1=np.sum(np.abs(guess)**2)
  guess*=support #Update real space object
  sos2=np.sum(np.abs(guess)**2) #Rescale amplitudes after update
  scale=(sos1/sos2)**0.5
  guess*=scale
  err.append(cerr)
  merr.append(scale)
#  print cerr, scale
  return guess


#Phase constrained HIO
def HIO(dmag,data,support,guess,err,merr):
  iguess=sf.fftn(guess)
  cerr=np.sum( ( np.abs(iguess)-np.abs(data) )**2 )/dmag
  phi=np.arctan2(iguess.imag,iguess.real)
  iguess=np.where(np.abs(data),data*np.exp(1.j*phi),0.0) #Replace with measured amplitudes
  oguess=guess
  guess=sf.ifftn(iguess)
  sos1=np.sum(np.abs(guess)**2)
  phr=np.arctan2(guess.imag, guess.real)
  chk=np.logical_and(support,np.logical_and((nph<phr),(phr<pph))) #Phase constraint
#  guess=support*guess+(1-support)*(oguess-p.beta*guess)
  guess=chk*guess+(1-chk)*(oguess-beta*guess) #Phase coonstrained HIO
  sos2=np.sum(np.abs(guess)**2) #Rescale amplitdues after update
  scale=(sos1/sos2)**0.5
  guess*=scale 
  err.append(cerr)
  merr.append(scale)
#  print cerr, scale
  return guess

#Shrink wrap and update support by scaling guassian smoothening with increasing iterations
def upsup(obj,i):
  ntot=n_iters/n_cycles #Only shrink support during the first cycle
  xo,xf=1.,ntot/severy #Calculate initial and final update points
  yo,yf=isig,float(fsig)
  i/=float(severy) #Count in number of updates
  if(xf==1):sig=(isig+fsig)/2. #0 division of only 1 update done
  else: sig=yo+(i-xo)*((yf-yo)/(xf-xo)) #Linearly scale sigma between initial and final values
  print "Real space sigma", sig
  rimage=np.abs(obj)

  smooth=np.abs(gauss_conv_fft(rimage,[sig,sig]))
  smooth/=smooth.max()
  supp=(smooth>=sfrac)*1 #Threshold as fraction of max
#  xyz_save(supp,'visuals/supp%d.xyz' %i)
  return supp


#based on Jesse Clarke
#save an fft by creating guassian in ft space
####################################################################
def gauss_conv_fft(arr, sigs):
  dims=arr.shape
  if len(sigs) != len(dims):
    return None
  tot=np.sum(arr)
  sigk=np.array(dims)/2.0/np.pi/sigs
  print sigk, "sigk"
  #gaussian needs to be wrap around to match the ft of arr
  gk=sf.fftshift(gauss(dims, sigk))

  arrk=sf.fftn(arr)
  convag=sf.ifftn(arrk*gk)
  convag=np.where(convag<0.0, 0.0, convag)
  convag*=tot/np.sum(convag)

  return convag

#dimensionally agnostic gaussian
####################################################################
def gauss(dims, sigs):
  if (len(sigs) != len(dims)):
    return None

  griddims=[]
  for d in dims:
    griddims.append( slice(-d/2+1,d/2+1) )
  grid=np.ogrid[ griddims ] 

  g=np.ones(dims) 
  d=0 
  for sig in sigs:
    if sig==0:
      g*=(grid[d]==0).astype(n.float)
    else:
      g*=np.exp(-0.5*grid[d]*grid[d]/sig**2)
    d+=1

  return g
```

```python
#Read image
im=tff.TiffFile(ifile)
imarr=im.asarray()
imarr=imarr.transpose(2,1,0) #I think the detector is rotated 90?
nx,ny,nz=imarr.shape[0],imarr.shape[1], imarr.shape[2]
print "Original number of X,Y and Z bins", nx, ny, nz

#Take only middle slice along Z
imarr=imarr[:,:,nz/2]
plt.imshow(np.log10(imarr))
```

```python

#Threshold values
imarr=np.where(imarr<thresh,0,imarr)

#Take square root of measured intensities
imarr=imarr**0.5

#Set size of object array
dx,dy=m.floor(dfs*nx),m.floor(dfs*ny)
dx,dy=int(dx+dx%2),int(dy+dy%2) #Numbers must be even
data=np.zeros((dx,dy))

#Find position of max intensity and centre data there
maxpos=np.unravel_index(imarr.argmax(),imarr.shape) #Gives you index of max as a tuple
shift_x=dx/2-maxpos[0]
shift_y=dy/2-maxpos[1]
data[shift_x:shift_x+nx,shift_y:shift_y+ny]=imarr
print "New bounds", shift_x,shift_x+nx,shift_y,shift_y+ny

data=sf.fftshift(data) #Shift data so D.C element is at 0
dmag=np.sum(np.abs(data)**2)
sos2=dmag/data.size #NOTE USE OF SIZE HERE BUT NOT AT SOS1
DC_elem=np.abs(data.ravel()[0])

#Build support
sx_pos,sx_neg=xf*dx/2,(-1)*xf*dx/2
sy_pos,sy_neg=yf*dy/2,(-1)*yf*dy/2
supportparams=((sx_pos,0,0),(sx_neg,0,0),(0,sy_pos,0),(0,sy_neg,0))
support=T.MakePoly(data.shape, supportparams) #Ross's function to create a polygon using bounds

#Create initial guess
rand1=np.random.random(support.shape)-0.5
guess=support*np.exp(1j*rand1)

#Normalize guess, normalize by energy
sos1=np.sum(np.abs(guess)**2)
print "Sum of real and fourier(averaged) squares", sos1, sos2
scale=(sos2/sos1)**0.5
guess*=scale
print "Scale, D.C element, Mag of sum", scale, DC_elem, np.abs(np.sum(guess))
print "New SoS guess and Sos data", np.sum(np.abs(guess)**2), sos2

iters=0
```

```python
err=[]
merr=[]
iters=0
for i in range(n_cycles):
 #First ER
 for k in range(n_ER):
   guess=ER(dmag,data[:,:],support[:,:],guess[:,:],err,merr)
   iters+=1
   if(iters%severy==0 and i==0): #Update support when it is time but only do it on first cycle
     support=upsup(guess[:,:],iters)

 #meth.xyz_save(guess,'visuals/ER.%d.xyz' %i)


 #Phase_HIO
 for k in range(n_HIO):
   guess=HIO(dmag,data[:,:],support[:,:],guess[:,:],err,merr)
   iters+=1
   if(iters%severy==0 and i==0):
     support=upsup(guess[:,:],iters)

 #meth.xyz_save(guess,'visuals/HIO.%d.xyz' %i)


 #Second ER
 for k in range(n_ER):
   guess=ER(dmag,data[:,:],support[:,:],guess[:,:],err,merr)
   iters+=1
   if(iters%severy==0 and i==0):
     support=upsup(guess[:,:],iters)
```
